# Country with the Highest Average Rapid Rating

```python
# Group the data by country and calculate the average rapid rating for each country
avg_rapid_ratings = df.groupby('Federation')['Rapid_rating'].mean().reset_index()

# Find the country with the highest average rapid rating
highest_avg_rapid_rating = avg_rapid_ratings[avg_rapid_ratings['Rapid_rating'] == avg_rapid_ratings['Rapid_rating'].max()]

# Print the result
print("Country with the Highest Average Rapid Rating:")
print(highest_avg_rapid_rating)
```

# Country with the maximum number of GMs 

```python
# Filter the dataset to only include players with the 'GM' title
gm_data = df[df['Title'] == 'GM']

# Group the data by 'Country' and count the number of GMs in each country
country_gm_counts = gm_data['Federation'].value_counts()

# Find the country with the maximum number of GMs
max_country = country_gm_counts.idxmax()
max_gm_count = country_gm_counts.max()

# Print the result
print(f"The country with the maximum number of GMs is {max_country} with {max_gm_count} GMs.")
```

# Each Country with it's highest rated player (Standard_Rating) 

```python

# Group the data by Federation and find the player with the highest rating in each group
highest_rated_players = df.groupby('Federation').apply(lambda x: x.loc[x['Standard_Rating'].idxmax()])

# Reset the index for the resulting DataFrame
highest_rated_players.reset_index(drop=True, inplace=True)

# Display the highest-rated player from each federation
print(highest_rated_players[['Federation', 'Name', 'Standard_Rating']])
```

```python
# Define a list of the top 7 federations (modify this list as needed)
top_7_federations = ['Country1', 'Country2', 'Country3', 'Country4', 'Country5', 'Country6', 'Country7']

# Filter the dataset to include only players from the top 7 federations
filtered_data = df[df['Federation'].isin(top_7_federations)]

# Find the player with the highest rating among the top 7 federations
highest_rated_player = filtered_data[filtered_data['Standard_Rating'] == filtered_data['Standard_Rating'].max()]

# Print the highest-rated player
print("Highest-rated player from the top 7 federations:")
print(highest_rated_player)
```

# Top 7 federations with the Highest Average Standard rating. 

```python

# Group the data by federation and calculate the average standard rating for each federation.
federation_ratings = df.groupby('Federation')['Standard_Rating'].mean().reset_index()

# Sort the federations by average standard rating in descending order.
top_federations = federation_ratings.sort_values(by='Standard_Rating', ascending=False)

# Display the top 7 federations with the highest average standard rating.
top_7_federations = top_federations.head(7)
print(top_7_federations)

filtered_df = df[df['Federation'].isin(top_7_federations)]

# Sort the filtered DataFrame by StandardRating in descending order
filtered_df = filtered_df.sort_values(by='Standard_Rating', ascending=False)

# Print the top players from the top 7 federations
print(filtered_df)
```
